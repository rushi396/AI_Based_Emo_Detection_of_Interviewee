# Total data prepration

```python
Total_Data_Path="../../Data/Main_Data/Images_Data"
```

```python
os.listdir("../../Data/Main_Data/Images_Data/")
```

```python
classes_list=os.listdir("../../Data/Main_Data/Images_Data/")
```

```python
classes_list
```

<h1 style="text-align:center;font-size:40px;">Preprocessing for Statastical Modelling</h1>

```python
for category in classes_list:
    print(category,len(list(os.listdir(os.path.join(Total_Data_Path,category)))))
```

```python
Total_Data=[]
class_value=0
for category in classes_list:
    class_path=os.path.join(Total_Data_Path,category)
#     print(class_path)
#     image_counter=0
    for image_name in os.listdir(class_path):
#         print(image_name)
        image_array=cv2.imread(os.path.join(class_path,image_name),cv2.IMREAD_GRAYSCALE)
#         print(image_array)
        image_array=cv2.resize(image_array,(120,120))
        modified_array=[]
        for row in list(image_array):
            modified_array=modified_array+list(row)
        Total_Data.append([modified_array,class_value])
#         if image_counter>439:
#             print(image_counter)
#             break
#         else:
#             image_counter+=1
#     break
    class_value+=1
    print(category)
```

```python
len(Total_Data)
```

```python
# Total_Data[0]
```

```python
random.shuffle(Total_Data)
```

```python
# Total_Data[0]
```

```python
Total_Features=[]
Total_Labels=[]
for data_point in Total_Data:
    Total_Features.append(data_point[0])
    Total_Labels.append(data_point[1]) 
```

```python
# Total_Features[0]
```

```python
set(Total_Labels)
```

```python
with open("../../Data/Preprocessed_Data/Images_Data/Total_Features.pickle","wb") as file:
    pickle.dump(Total_Features,file)
with open("../../Data/Preprocessed_Data/Images_Data/Total_Labels.pickle","wb") as file:
    pickle.dump(Total_Labels,file)
```

<h1 style="text-align:center;font-size:40px;">Preprocessing for Neural Modelling</h1>

```python
Total_Data=[]
class_value=0
for category in classes_list:
    class_path=os.path.join(Total_Data_Path,category)
#     print(class_path)
    image_counter=0
    for image_name in os.listdir(class_path):
#         print(image_name)
        image_array=cv2.imread(os.path.join(class_path,image_name),cv2.IMREAD_GRAYSCALE)
#         print(image_array)
        image_array=cv2.resize(image_array,(120,120))
        Total_Data.append([image_array,class_value])
        if image_counter>400:
            print(image_counter)
            break
        else:
            image_counter+=1
#     break
    class_value+=1
    print(category)
random.shuffle(Total_Data)
```

```python
Total_Features=[]
Total_Labels=[]
for data_point in Total_Data:
    Total_Features.append(data_point[0])
    Total_Labels.append(data_point[1]) 
```

```python
with open("../../Data/Preprocessed_Data/Images_Data/CNN_Total_Features.pickle","wb") as file:
    pickle.dump(Total_Features,file)
with open("../../Data/Preprocessed_Data/Images_Data/CNN_Total_Labels.pickle","wb") as file:
    pickle.dump(Total_Labels,file)
```
