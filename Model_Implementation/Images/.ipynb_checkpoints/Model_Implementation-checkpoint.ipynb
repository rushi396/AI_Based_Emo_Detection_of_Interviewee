<h1 style="text-align:center;font-size:30px;">Input Preparation Section</h1>

```python
def getPreprocessedInput(filename):
    image_array=cv2.imread(filename,cv2.IMREAD_GRAYSCALE)
    image_array=cv2.resize(image_array,(120,120))
    modified_array=[]
    for row in list(image_array):
        modified_array=modified_array+list(row)
    return modified_array
```

```python
input_array=getPreprocessedInput(filename)
```

<h1 style="text-align:center;font-size:40px;">Neural Network Approach</h1>

```python
with open("../../Data/Preprocessed_Data/Images_Data/CNN_Total_Features.pickle","rb") as file:
    Total_Features=pickle.load(file)
with open("../../Data/Preprocessed_Data/Images_Data/CNN_Total_Labels.pickle","rb") as file:
    Total_Labels=pickle.load(file)
training_count=int(len(Total_Features)*0.9)
```

```python
Training_Features=Total_Features[:training_count]
Training_Labels=Total_Labels[:training_count]
Testing_Features=Total_Features[training_count:]
Testing_Labels=Total_Labels[training_count:]
```

```python
def getCNNPreprocessedInput(filename):
    image_array=cv2.imread(filename,cv2.IMREAD_GRAYSCALE)
    image_array=cv2.resize(image_array,(120,120))
    image_array=numpy.expand_dims(numpy.array([image_array]),axis=-1)
    return image_array
```

```python
def getClass(array):
    index=0
    max_=max(array)
    for i in range(len(array)):
        if array[i]==max_:
            index=i
#     print(class_list[index])
    return class_list[index]
```

```python
CNN_input_array=getCNNPreprocessedInput(filename)
```

```python
Training_Features=numpy.array(Training_Features)
Testing_Features=numpy.array(Testing_Features)
```

```python
Training_Features=numpy.expand_dims(Training_Features,axis=-1)
Testing_Features=numpy.expand_dims(Testing_Features,axis=-1)
```

```python
Training_Features.shape
```

```python
set(list(Testing_Labels))
```

```python
Simple_Model_1=keras.models.Sequential(
     [
         keras.layers.Conv2D(filters=64,kernel_size=3,strides=(1,1),padding='valid',activation='relu',input_shape=[120,120,1]),
         keras.layers.MaxPool2D(pool_size=(2,2)),
         keras.layers.Flatten(),
         keras.layers.Dense(units=256,activation='relu'),
         keras.layers.Dense(units=64,activation='relu'),
         keras.layers.Dense(units=len(class_list),activation='softmax'),
     ]
 )
```

```python
Simple_Model_1.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
Simple_Model_1.fit(Training_Features,numpy.array(Training_Labels),epochs=10,verbose=1)
```

```python
Simple_Model_1.save("../../Models/Images/Simple_Model_for_Emotion_images.h5") 
```
